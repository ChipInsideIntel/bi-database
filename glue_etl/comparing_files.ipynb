```python
import pandas as pd
```

```python
df_glue = pd.read_parquet('/home/cowmed14/Downloads/users_active.parquet')

df_pandas = pd.read_csv('/home/cowmed14/Downloads/OneDrive_1_11-14-2025/users_active.csv')

df_athena = pd.read_csv('/home/cowmed14/Downloads/USERS_ACTIVE_QUERY_ATHENA.csv')
```

```python
display(df_glue.shape)
print(df_glue.columns)

display(df_pandas.shape)
print(df_pandas.columns)

display(df_athena.shape)
print(df_athena.columns)
```

```python
#Verificar colunas que possuem no pandas e nao possuem no athena e no glue

set_pandas = set(df_pandas.columns)
set_athena = set(df_athena.columns)
set_glue = set(df_glue.columns)

print("Colunas no pandas mas nao no athena:", set_pandas - set_athena)
print("Colunas no pandas mas nao no glue:", set_pandas - set_glue)

print("Colunas no athena mas nao no pandas:", set_athena - set_pandas)
print("Colunas no glue mas nao no pandas:", set_glue - set_pandas)

#COlunas que estao no glue e nao estao no athena
print("Colunas no glue mas nao no athena:", set_glue - set_athena)
#Colunas que estao no athena e nao estao no glue
print("Colunas no athena mas nao no glue:", set_athena - set_glue)
```
